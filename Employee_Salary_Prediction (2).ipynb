```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Dictionary to hold all models and their predictions
models = {
    "Decision Tree": classifier.predict(x_test),
    "Random Forest": classifier1.predict(x_test),
    "Logistic Regression": logreg.predict(x_test),
    "Gaussian NB": gnb.predict(x_test),
    "Bernoulli NB": bnb.predict(x_test),
    "Multinomial NB": mnb.predict(X_test_mnb) ,
    "Support Vector Machine(SVM)" :model.predict(X_test_scaled) ,
    "KNeighbors Classifier":knn.predict(x_test),
    "MLP Classifier":mlp.predict(x_test),
    "Gradient Boosting Classifier":gbc.predict(x_test),
    "Ensemble-Voting Classifier":voting.predict(x_test),
    "Grid Search CV":grid.predict(x_test)
}

# Create a comparison DataFrame
comparison = []

for model_name, predictions in models.items():
    # For Multinomial NB, predictions are already in string format and based on X_test_mnb
    if model_name == "Multinomial NB":
        current_y_test = y_test_unscaled # Use the unscaled y_test for Multinomial NB
        current_predictions = predictions
    else:
        # Use the original y_test and predictions for other models
        current_y_test = y_test
        current_predictions = predictions


    acc = accuracy_score(current_y_test, current_predictions)
    # Use the original string labels for metric calculations
    # Specify positive label for binary classification
    prec = precision_score(current_y_test, current_predictions, pos_label='>50K')
    rec = recall_score(current_y_test, current_predictions, pos_label='>50K')
    f1 = f1_score(current_y_test, current_predictions, pos_label='>50K')

    comparison.append({
        "Model": model_name,
        "Accuracy": round(acc, 4),
        "Precision": round(prec, 4),
        "Recall": round(rec, 4),
        "F1 Score": round(f1, 4)
    })

comparison_df = pd.DataFrame(comparison).sort_values(by="F1 Score", ascending=False)
print(comparison_df)
print("\nBest Classification Model:", comparison_df.iloc[0]["Model"])
```
